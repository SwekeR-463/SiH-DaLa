```python
import pandas as pd
import numpy as np
```

```python
stats = {
    'Payload (Ton)': {'mean': 145.205, 'std': 17.082, 'min': 77.800, 'max': 184.900},
    'Empty Travel Distance (km)': {'mean': 2.280, 'std': 0.842, 'min': 1.000, 'max': 4.700},
    'Load Travel Distance (km)': {'mean': 2.307, 'std': 0.853, 'min': 0.800, 'max': 4.300},
    'Empty Speed (km/hr)': {'mean': 20.922, 'std': 3.705, 'min': 13.091, 'max': 31.086},
    'Load Speed (km/hr)': {'mean': 20.721, 'std': 3.472, 'min': 10.655, 'max': 30.000},
    'Empty Travel Time (hr)': {'mean': 0.108, 'std': 0.031, 'min': 0.053, 'max': 0.199},
    'Empty Stop Time (hr)': {'mean': 0.055, 'std': 0.055, 'min': 0.001, 'max': 0.369},
    'Load Time (hr)': {'mean': 0.048, 'std': 0.015, 'min': 0.026, 'max': 0.065},
    'Load Stop Time (hr)': {'mean': 0.017, 'std': 0.008, 'min': 0.009, 'max': 0.066},
    'Load Travel Time (hr)': {'mean': 0.111, 'std': 0.034, 'min': 0.052, 'max': 0.198},
    'TKPH (Ton.Km/Hr)': {'mean': 144.782, 'std': 107.440, 'min': 28.183, 'max': 439.721},
}
```

```python
np.random.seed(42)
data = {}
for column, stat in stats.items():
    data[column] = np.clip(
        np.random.normal(loc=stat['mean'], scale=stat['std'], size=1000),
        stat['min'], stat['max']
    )
```

```python
synthetic_df = pd.DataFrame(data)
```
